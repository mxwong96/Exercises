```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Basic Graph: https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
# for line 1
x=[1,2,3]
y=[2,4,6]

# resize the graph
plt.figure(figsize=(5,3), dpi=200)

#for line 2
x2 = np.arange(0,5,0.5)
y2 = x2**2 #square

# plot part of the graph as line
plt.plot(x2[:6],y2[:6], 'r', label='x2')  # [:6] takes the first 5 values
# plot remainder of the graph
plt.plot(x2[5:],y2[5:], 'r--') # format to plot after 5th value

plt.plot(x,y, 'b^--', label='2x')
plt.xlabel("X Axis")
plt.xticks([0,1,2,3,4,5])
plt.ylabel("Y Axis")
plt.yticks(range(0,21))
plt.title("First Graph", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20, 'color': 'g'})

plt.legend()

plt.savefig('mygraph.png', dpi=300)  #save graph into current directory

plt.show()
```

```python
# First bar chart

labels = ['A', 'B', 'C']
values = [1,4,2]


# we can set up different patterns for each bar on the chart by assignning the chart itself as an object
bars = plt.bar(labels, values)
patterns = ['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    

print(type(bars))
print(bars[0])

plt.bar(labels, values)
plt.show()
```
